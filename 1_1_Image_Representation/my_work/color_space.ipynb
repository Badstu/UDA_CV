```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
%matplotlib inline
image = cv2.imread('../images/water_balloons.jpg')
image_copy = np.copy(image)
image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
r, g, b = image_copy[:, :, 0], image_copy[:, :, 1], image_copy[:, :, 2]
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title('RED')
ax1.imshow(r, cmap='gray')
ax2.set_title('GREEN')
ax2.imshow(g, cmap='gray')
ax3.set_title('BLUE')
ax3.imshow(b, cmap='gray')
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = hsv[:, :, 0], hsv[:, :, 1], hsv[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')
ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')
ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```python
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])

lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
```

```python
mask_rgb = cv2.inRange(image, lower_pink, upper_pink) #在范围之内的保留，范围之外的=0
masked_imgae_rgb = np.copy(image)
masked_imgae_rgb[mask_rgb == 0] = [0, 0, 0]
plt.imshow(masked_imgae_rgb)
```

```python
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)
masked_image_hsv = np.copy(image)
masked_image_hsv[mask_hsv == 0] = 0
plt.imshow(masked_image_hsv)
```

```python
image = mpimg.imread('../images/car_green_screen2.jpg')
plt.imshow(image)
```

```python
lower_green = np.array([0, 80, 0])
upper_green = np.array([100, 255, 100])
mask = cv2.inRange(image, lower_green, upper_green)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
image_copy = np.copy(image)
image_copy[mask != 0] = [0, 0, 0]
plt.imshow(image_copy)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = hsv[:, :, 0], hsv[:, :, 1], hsv[:, :, 2]

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title('H channel')
ax1.imshow(h, cmap='gray')
ax2.set_title('S channel')
ax2.imshow(s, cmap='gray')
ax3.set_title('V channel')
ax3.imshow(v, cmap='gray')
```

```python
lower_hue = np.array([20, 0, 0])
upper_hue = np.array([80, 255, 255])
```

```python
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)
plt.imshow(mask_hsv, cmap='gray')
```

```python
masked_image_hsv = np.copy(image)
masked_image_hsv[mask_hsv != 0] = 0
plt.imshow(masked_image_hsv)
```
