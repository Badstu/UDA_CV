```python
def standardize_input(image):
    standard_im = cv2.resize(image, (1100, 600))
    return standard_im
```

```python
def encode(label):
    numerical_val = 1 if label == "day" else 0
    return numerical_val
```

```python
array_dev = np.array(IMAGE_LIST)
array_dev.shape
```

```python
def standardize(image_list):
    standard_list = []
    for item in image_list:
        image = item[0]
        label = item[1]
        
        standardized_im = standardize_input(image)
        binary_label = encode(label)
        standard_list.append((standardized_im, binary_label))
    return standard_list

STANDARDIZED_LIST = standardize(IMAGE_LIST)
```

```python
image_num = 2
test_im = STANDARDIZED_LIST[image_num][0]
test_label = STANDARDIZED_LIST[image_num][1]

hsv = cv2.cvtColor(test_im, cv2.COLOR_RGB2HSV)
print('label: {}'.format(test_label))
h = hsv[:, :, 0]
s = hsv[:, :, 1]
v = hsv[:, :, 2]

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20, 10))
ax1.set_title('Standardized image')
ax1.imshow(test_im)
ax2.imshow(h, cmap='gray')
ax3.imshow(s, cmap='gray')
ax4.imshow(v, cmap='gray')
```

```python
def avg_brightness(rgb_image):
    hsv = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)
    sum_brightness = np.sum(hsv[:, :, 2])
    avg = sum_brightness / (600*1100.0)
    return avg
```

```python
image_num = 161
test_im = STANDARDIZED_LIST[image_num][0]
avg = avg_brightness(test_im)
print(avg)
plt.imshow(test_im)
```

```python
def estimate_label(rgb_image):
    avg = avg_brightness(rgb_image)
    if avg > 100:
        predicted_label = 1
    else:
        predicted_label = 0
    return predicted_label
```

```python
TEST_LIST = helpers.load_dataset(image_dir_test)
TEST_STANDARDIZED_LIST = standardize(TEST_LIST)
random.shuffle(TEST_STANDARDIZED_LIST)
```

```python
def get_misclassified_images(test_images):
    misclassification_list = []
    for item in TEST_STANDARDIZED_LIST:
        test_im = item[0]
        test_label = item[1]
        predicted_label = estimate_label(test_im)

        if predicted_label != test_label:
            misclassification_list.append((test_im, test_label, predicted_label))
    
    return misclassification_list
```

```python
MISCLASSIFIED = get_misclassified_images(TEST_STANDARDIZED_LIST)
total = len(TEST_STANDARDIZED_LIST)
num_correct = total - len(MISCLASSIFIED)
accuracy = num_correct / total
print('Accuracy: {}'.format(accuracy))
print('Number of misclassified images = {} out of {}'
      .format(len(MISCLASSIFIED), total))
```
