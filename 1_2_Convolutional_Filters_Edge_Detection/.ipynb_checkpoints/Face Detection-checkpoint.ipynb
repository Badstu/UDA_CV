```python
from __future__ import print_function, division
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('../data/1_2/images/multi_faces.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(20, 10))
plt.imshow(gray, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, 4, 6)
```

```python
print(len(faces))
print(faces)
```

```python
img_with_detections = np.copy(image)

for (x, y, w, h) in faces:
    cv2.rectangle(img_with_detections, (x, y), (x+w, y+h), (255, 0, 0), 5)

plt.figure(figsize=(20, 10))
plt.imshow(img_with_detections)
```

```python
eye_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_eye.xml')
eyes = eye_cascade.detectMultiScale(gray, 1.8, 2)
print(len(eyes))
```

```python
img_with_eyes = np.copy(image)
for (x, y, w, h) in eyes:
    cv2.rectangle(img_with_eyes, (x, y), (x+w, y+h), (255, 0, 0), 3)

plt.figure(figsize=(20, 10))
plt.imshow(img_with_eyes) 
```
