```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/monarch.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
pixle_vals = image.reshape((-1, 3))
pixle_vals = np.float32(pixle_vals)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
criteria
```

```python
k = 6
retval, labels, centers = cv2.kmeans(pixle_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
centers = np.uint8(centers)
segment_data = centers[labels.flatten()]
```

```python
labels.shape
```

```python
segment_image = segment_data.reshape((image.shape))
labels_reshape = labels.reshape(image.shape[0], image.shape[1])
plt.imshow(segment_image)
```

```python
plt.imshow(labels_reshape==5, cmap='gray')
```

```python
cluster = 4
masked_image = np.copy(image)
masked_image[labels_reshape == cluster] = [0, 255, 0]
plt.imshow(masked_image)
```
